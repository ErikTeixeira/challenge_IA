```python
# NOTE: Your prompt contains media inputs that are not directly supported by the
# Gemini Files API. Preprocessing will be required for these inputs. Specific
# information is provided below.

"""
Install the Google AI Python SDK

$ pip install google-generativeai

See the getting started guide for more information:
https://ai.google.dev/gemini-api/docs/get-started/python
"""

import os
import pandas as pd
import google.generativeai as genai

# Defina sua chave de API aqui
api_key = "xxxxxxxxxxxxxxxx"

genai.configure(api_key=api_key)


# Create the model
# See https://ai.google.dev/api/python/google/generativeai/GenerativeModel
generation_config = {
  "temperature": 1,
  "top_p": 0.95,
  "top_k": 64,
  "max_output_tokens": 8192,
  "response_mime_type": "text/plain",
}
safety_settings = [
  {
    "category": "HARM_CATEGORY_HARASSMENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_HATE_SPEECH",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_SEXUALLY_EXPLICIT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
  {
    "category": "HARM_CATEGORY_DANGEROUS_CONTENT",
    "threshold": "BLOCK_MEDIUM_AND_ABOVE",
  },
]

model = genai.GenerativeModel(
  model_name="gemini-1.5-pro-latest",
  safety_settings=safety_settings,
  generation_config=generation_config,
)


# Function to read and extract data from Excel files
def extract_excel_data(file_path):
    df = pd.read_excel(file_path)
    return df


# Extract file contents
file_paths = [
    "/content/reviews0.2.xlsx",
    "/content/reviews0.3.xlsx",
    "/content/reviews0.4.xlsx",
    "/content/reviews0.5.xlsx",
    "/content/reviews0.6.xlsx"
]

# Extract data from each file
extracted_data = []
for file_path in file_paths:
    data = extract_excel_data(file_path)
    extracted_data.append(data)

# Combine all extracted data into a single dataframe
combined_data = pd.concat(extracted_data, ignore_index=True)

# Convert the combined data to a string format to use as input for the model
data_str = combined_data.to_string()



# Create a chat session and send a message with the extracted data
chat_session = model.start_chat(
  history=[
    {
      "role": "user",
      "parts": [
        "me fale quais s√£o os 15 lugares que tem reviews nestes arquivos",
        data_str  # Pass the extracted data as part of the prompt
      ],
    },
    {
      "role": "model",
      "parts": [
        "Os 15 lugares com reviews nos seus arquivos s√£o:\n\n1. Museu da Imagem e do Som (MIS)\n2. Museu de Arte de S√£o Paulo Assis Chateaubriand\n3. Museu Pinacoteca de S√£o Paulo\n4. Ibirapuera Park\n5. Parque Villa-Lobos\n6. Parque da √Ågua Branca\n7. Teatro Municipal de S√£o Paulo\n8. Teatro S√©rgio Cardoso\n9. Teatro Proc√≥pio Ferreira\n10. Duas Terezas\n11. Petros Greek Taverna\n12. Merenda da Cidade\n13. JazzB\n14. Bar Brahma Centro\n15. Clube do Minhoca \n",
      ],
    },
    {
      "role": "user",
      "parts": [
        "que perguntas podem ser feitar apartir dos dados que tem nestes arquivos para que voce consiga gerar um itinerario da cidade de s√£o paulo utilizando os dados do arquivos que eu te dei?",
      ],
    },
    {
      "role": "model",
      "parts": [
        "Para gerar um itiner√°rio personalizado para voc√™ em S√£o Paulo, utilizando os dados dos arquivos, preciso entender melhor suas prefer√™ncias e vontades! Que tal responder a estas perguntas? üòä\n\n**Sobre voc√™:**\n\n* **Quantos dias voc√™ ter√° em S√£o Paulo?** \n* **Voc√™ prefere um roteiro mais focado em arte e cultura, gastronomia, vida noturna, atividades ao ar livre ou um mix de tudo?**\n* **Voc√™ estar√° viajando sozinho, em casal, com amigos ou fam√≠lia?** \n* **Qual o seu or√ßamento para a viagem?** (Isso ajudar√° na escolha de restaurantes, acomoda√ß√µes e atividades)\n* **Quais s√£o seus principais interesses?** (Ex: fotografia, hist√≥ria, arquitetura, m√∫sica)\n* **H√° algo que voc√™ definitivamente N√ÉO quer fazer?** (Ex: visitar zool√≥gicos, ir a shows)\n* **Quais meios de transporte voc√™ pretende usar?** (Ex: transporte p√∫blico, carro, bicicleta)\n* **Voc√™ tem alguma restri√ß√£o alimentar ou necessidade especial de acessibilidade?\n\n**Com base nas suas respostas, posso criar um itiner√°rio incr√≠vel para voc√™, explorando os lugares com as melhores avalia√ß√µes e que se encaixam no seu perfil!** üòâ \n",
      ],
    },
  ]
)



response = chat_session.send_message("eu tenho 2 dias em s√£o paulo, eu quero um mix de tudo, eu estou viajando sozinho, me or√ßamento da viagem √© 5000 reais, eu pretendo usar carro enquanto estiver em s√£o paulo e n√£o tenho nenhuma restri√ß√£o alimentar")

print(response.text)
print(chat_session.history)
```
